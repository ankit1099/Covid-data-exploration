```python
# Import libraries 
import numpy as np
import pandas as pd
import seaborn as sns

import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
plt.rcParams['figure.figsize'] = 12,8  # adjusts the configuration of the plots


import datetime
from datetime import datetime
from dateutil.parser import parse

# Read in the data

df = pd.read_csv('Downloaded_datas\movies.csv')
```

```python
# Let's look at data

df.head()
```

```python
# check for missing data

df.isnull().any()
```

```python
# dropping rows having missing data

df.dropna(inplace=True)
```

```python
# checking again for null values

df.isnull().any()
```

```python
# Data types of the columns

df.dtypes
```

```python
# change data types of columns

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')
```

```python
df.head()
```

```python
# create correct year column

df['yearCorrect'] = df['released'].astype('str').apply(lambda x: parse(x.split('(')[0]).year)
```

```python
df.head()
```

```python
df.sort_values(by='gross',inplace=True,ascending=False)
```

```python
pd.set_option('display.max_rows',None)
```

```python
# Check list of Companies
df['company'].value_counts().sort_index()
```

```python
df.head()
```

```python
# Budget vs Gross Revenue

plt.scatter(x=df['budget'],y=df['gross'])

plt.title('Budget vs Gross Earning')

plt.xlabel('Budget')
plt.ylabel('Gross Earning')
# company vs Revenue
```

```python
sns.regplot(x='budget',y='gross',data=df,scatter_kws={'color':'red'},line_kws={'color':'blue'})
```

```python
# looking for correlation
```

```python
df.corr(method='pearson')
```

```python
# High correlation between budget and gross
```

```python
correlation_matrix = df.corr(method='pearson',)

sns.heatmap(correlation_matrix,annot=True)

plt.title('correlation Matrix')
plt.xlabel('Movie Feature')
plt.ylabel('Movie Feature')
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized)[col_name].dtype == 'object':
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized.head()
```

```python
correlation_matrix = df_numerized.corr(method='pearson',)

sns.heatmap(correlation_matrix,annot=True)

plt.title('correlation Matrix')
plt.xlabel('Movie Feature')
plt.ylabel('Movie Feature')
```

```python
corr_numerized = df_numerized.corr()

corr_pairs = corr_numerized.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[sorted_pairs>0.5]
high_corr
```
